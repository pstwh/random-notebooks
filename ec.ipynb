```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import spline
```

```python
def smooth(x, y):
    
    xx = np.array(x)
    sx = np.argsort(xx)
    xx = xx[sx]
    yy = np.array(y)
    yy = yy[sx]
    
    x_new = np.linspace(xx.min(), xx.max(), 300)
    y_new = spline(xx, yy,x_new)
    
    return x_new, y_new
```

```python
preco_unidade = np.array([200, 100, 50, 25, 10])
ana = [0, 2, 3, 5, 7]
joao = [1, 3, 5, 7, 10]
carlos = [0, 2, 4, 6, 10]
maria = [3, 5, 8, 10, 13]

thiago = [12, 9, 7, 3, 0]
gabriel = [15, 8, 5, 1, 0]
joana = [13, 10, 8, 4, 0]
```

```python
consumidores = [ana, joao, carlos, maria]
produtores = [thiago, gabriel, joana]
total = consumidores + produtores
```

```python
sum_consumidores = np.array(np.sum(consumidores,axis=0).tolist())
sum_produtores = np.array(np.sum(produtores,axis=0).tolist())
```

```python
plt.ylabel("Preço")
plt.xlabel("Quantidade")

print(sum_produtores)
print(sum_consumidores)

x1, y1 = smooth(sum_consumidores, preco_unidade)
x2, y2 = smooth(sum_produtores, preco_unidade)

x_ana, y_ana = smooth(ana, preco_unidade)
x_joao, y_joao = smooth(joao, preco_unidade)
x_carlos, y_carlos = smooth(carlos, preco_unidade)
x_maria, y_maria = smooth(maria, preco_unidade)

x_thiago, y_thiago = smooth(thiago, preco_unidade)
x_gabriel, y_gabriel = smooth(gabriel, preco_unidade)
x_joana, y_joana = smooth(joana, preco_unidade)

plt.plot(x1, y1, '--', label='Demanda')
plt.plot(x2, y2, '-', label='Oferta')
plt.title("Oferta e Demanda")
plt.grid(True)
plt.plot(13.2, 30, 'o', color='green')
plt.plot(25.8, 30, 'o', color='green')
plt.plot([13.2, 25.8], [30, 30], ':', color='green')
plt.rc('grid', linestyle="--", color='#eeeeee')
plt.legend()
```
